```python
from qiskit import QuantumCircuit, execute, Aer
from math import sqrt, pi
from qiskit.visualization import plot_histogram, plot_bloch_multivector
sim = Aer.get_backend('aer_simulator')
```

```python
# Pauli Gates
## 1.1 X-Gate
```

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw()
```

```python
qc.save_statevector()
state = execute(qc, sim).result().get_statevector()
plot_bloch_multivector(state) # 상태가 완전히 바뀜을 알 수 있음
```

```python
## 1.2 Y-Gate & Z-Gate
qc = QuantumCircuit(1)
initial_state = [0, 1]
qc.initialize(initial_state, 0)

qc.z(0)
qc.draw()
```

```python
qc.save_statevector()
result = execute(qc, sim).result().get_statevector()
plot_bloch_multivector(result)
```

```python
## 1-(2) |1>을 하다마드 게이트에 통과시켰을 때의 결과는?
qc = QuantumCircuit(1)
initial_state = [0, 1]
qc.initialize(initial_state, 0)
qc.h(0)
qc.draw()

qc.save_statevector()
result = execute(qc, sim).result().get_statevector()
plot_bloch_multivector(result)
```

```python
## 1-(3) |+>을 하다마드 게이트에 통과시켰을 때의 결과는?
qc = QuantumCircuit(1)
initial_state = [1/sqrt(2), 1/sqrt(2)]
qc.initialize(initial_state, 0)
qc.h(0)
qc.draw()

qc.save_statevector()
result = execute(qc, sim).result().get_statevector()
plot_bloch_multivector(result)
```
