```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python
affected = pd.read_excel("확진자 수 데이터.xlsx")
mobility = pd.read_excel("구글 mobility data.xlsx")
```

```python
mobility_df = mobility.iloc[:700,[1,7,8,9,10,11,12]]
```

```python
for i in range(695):
    for j in range(7):
        mobility_df.iloc[i,j] = mobility_df.iloc[i:i+7,j].mean()

mobility_df
```

```python
import matplotlib.pyplot as plt
for i in range(1,7):
    fig, ax = plt.subplots()
    ax.plot(mobility_df[mobility_df.columns[i]], label = mobility_df.columns[i], color = 'blue')
   
    plt.xticks(range(701))
    plt.legend()

    ax1 = ax.twinx()
    ax1.plot(mobility_df['count'], color = 'red', label = 'infected')
    plt.legend()
    plt.show()
```
