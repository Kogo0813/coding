```python
mobility = pd.read_excel('../data/mobility.xlsx')
mobility
```

```python
mobility.set_index('Date', inplace = True)
```

```python
seoul_mob = mobility.columns[mobility.columns.str.contains('seoul')]
metro_mob = mobility.columns[mobility.columns.str.contains('metro')]
```

```python
mobility_seoul = mobility.loc[:, seoul_mob]
mobility_metro = mobility.loc[:, metro_mob]
```

```python
mobility_seoul['total'] = mobility_seoul.sum(axis = 1)
mobility_metro['total'] = mobility_metro.sum(axis = 1)
```

```python
mobility_seoul['total'].plot()
mobility_metro['total'].plot()
```

```python
mobility_seoul.resample('M')['total'].sum().plot()
mobility_metro.resample('M')['total'].sum().plot()
```

```python
infect = pd.read_excel('../data/COVID19_variants_국내.xlsx')
infect
```

```python
infect.rename(columns = {'Unnamed: 0' : 'date'}, inplace = True)
```

```python
infect['total'] = infect.sum(axis = 1)
```

```python
infect['date'] = pd.to_datetime(infect['date'])
```

```python
pivot_infect = infect.pivot_table(index = 'date', columns= 'city', values = 'total', aggfunc= 'sum')
```

```python
pivot_infect = pivot_infect.resample('M').sum()
```

```python
pivot_infect.to_csv('../data/pivot_infect.csv')
```

```python
city_3 = pivot_infect.loc[:, ['Seoul', 'Busan', 'Daegu']]
```
